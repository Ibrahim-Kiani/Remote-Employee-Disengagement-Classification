### 1.2 Target variable comparison 
***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------------
# 1) PREPARE THE DATA
# -------------------------------
numerics = [
    'is_disengaging', 'average_meeting_hours', 'code_commits_per_week',
    'messages_sent_slack', 'screen_active_minutes',
    'sentiment_score', 'timezone_alignment'
]

group_means = (
    df[numerics]
    .groupby('is_disengaging')
    .mean()
    .T
)

normalized = group_means.div(group_means.max(axis=1), axis=0) * 100
normalized.index.name = "Feature"

plot_df = (
    normalized
    .reset_index()
    .melt(id_vars="Feature", var_name="Disengaged", value_name="Percent")
)
plot_df["Disengaged"] = plot_df["Disengaged"].astype("category")

# -------------------------------
# 2) PLOT WITH SEABORN
# -------------------------------
sns.set_theme(style="whitegrid", context="talk", font_scale=0.95)

palette = ["white", "#1f77b4"]

fig, ax = plt.subplots(figsize=(12, 6), facecolor='white')
ax.set_facecolor("#f8f9fa")  # light background for plot area

sns.barplot(
    data=plot_df,
    x="Feature", y="Percent",
    hue="Disengaged", palette=palette,
    edgecolor="black", linewidth=1.5,
    ax=ax
)

# Value labels
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f"{height:.1f}%",
                    (p.get_x() + p.get_width() / 2, height),
                    ha="center", va="bottom",
                    fontsize=9)

# Styling
ax.set_title("Relative Feature Means by Disengagement Status", pad=30)
ax.set_ylabel("Mean (% of Feature Max)")
ax.set_xlabel("")
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="center")

# Move legend to upper right corner with offset
ax.legend(title="Disengaged", loc="upper left", bbox_to_anchor=(1.01, 1), borderaxespad=0)

sns.despine(ax=ax, top=True, right=True)

plt.tight_layout()
plt.show()
```

#### 1. Comparing the mean numerical features of both engaged and disengaged employees shows that disengaged employees are on average less active than their counterparts.
#### 2. There seems to be a higher correlation between timezone alignment and employee disengagement.
#### 3. Higher sentiment score seems to be correlated with more disengagement.

# 2. Machine learning
***

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    classification_report, confusion_matrix
)
import statsmodels.api as sm
```

### 2.1 Feature engineering
***

```python
weekly = ['average_meeting_hours', 'code_commits_per_week', 'messages_sent_slack', 'screen_active_minutes', 'sentiment_score']
# Aggregate based on weekly data's mean
grouped = df.groupby('employee_id', as_index=False)[weekly].agg(['mean'])
grouped.columns = ['employee_id'] + [f"{col[0]}_{col[1]}" for col in grouped.columns[1:]]
df = pd.merge(grouped,df,on='employee_id',how='inner')
col = df.columns.to_list()
col.remove('employee_id')
col.remove('week_number')
print(col)
df
```

#### Grouped employee by averages throughout all weekly metrics.
---
### 2.2 Data Transformation
---

```python

pivoted_dfs = []

for metric in weekly:
    # Create a subset with only the metric we're pivoting
    subset = df[['employee_id', 'week_number', metric]].copy()
    
    # Pivot the metric
    pivoted = subset.pivot(index='employee_id', columns='week_number', values=metric)
    
    # Rename columns to include metric name
    pivoted.columns = [f'{metric}_week_{int(col)}' for col in pivoted.columns]
    
    pivoted_dfs.append(pivoted)

# Combine all pivoted metrics
df2 = pd.concat(pivoted_dfs, axis=1)
static_cols = df.groupby('employee_id')[col].first()
df2 = pd.concat([df2, static_cols], axis=1)
df2 = df2.reset_index()
print(df2.columns)
df2
```

#### Pivoted the week number based on employee id to convert the 6 rows per employee format to one row per employee.
***
### 2.3 Machine Learning
***

```python
X = df2.drop(columns=['employee_id', 'is_disengaging'])
y = df2['is_disengaging']
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# 2️⃣ Build & fit pipeline
pipe = Pipeline([
    ("scale", StandardScaler()),
    ("select", SelectFromModel(
        LogisticRegression(
            penalty="l1",
            solver="saga",
            C=1.0,
            max_iter=5000,
            n_jobs=-1
        ),
        threshold="mean"
    )),
    ("clf", LogisticRegression(
        penalty="l2",
        solver="lbfgs",
        C=1.0,
        max_iter=5000
    ))
])
pipe.fit(X_train, y_train)

# 3️⃣ Predict & evaluate
y_pred  = pipe.predict(X_test)
y_prob  = pipe.predict_proba(X_test)[:, 1]

print(f"Accuracy : {accuracy_score(y_test, y_pred):.3f}")
print(f"Precision: {precision_score(y_test, y_pred):.3f}")
print(f"Recall   : {recall_score(y_test, y_pred):.3f}")
print(f"F1-score : {f1_score(y_test, y_pred):.3f}")
print(f"ROC-AUC  : {roc_auc_score(y_test, y_prob):.3f}\n")

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred), "\n")

# 4️⃣ Extract selected features & scikit‑learn coefficients
sel_mask      = pipe.named_steps["select"].get_support()
sel_features  = X_train.columns[sel_mask]
sk_coefs      = pipe.named_steps["clf"].coef_.ravel()

coef_df = pd.DataFrame({
    "feature": sel_features,
    "sk_coef": sk_coefs
}).sort_values("sk_coef", ascending=False).reset_index(drop=True)

print("Scikit‑learn coefficients:")
print(coef_df)
print(f"\nIntercept (β₀): {pipe.named_steps['clf'].intercept_[0]:.4f}\n")

# 5️⃣ Re‑fit in statsmodels for p‑values & CIs
#    → use the exact scaled & selected design matrix
X_train_scaled = pipe.named_steps["scale"].transform(X_train)
X_train_sel    = X_train_scaled[:, sel_mask]

# add constant (intercept)
X_train_sm = sm.add_constant(X_train_sel, prepend=False)

logit_sm = sm.Logit(y_train, X_train_sm).fit(disp=False)
print("Statsmodels Logistic Regression Results:")
print(logit_sm.summary())
```

#### Perfect classification on all metrics. This suggests data leakage or artificial data. After careful evaluation, the data is judged to be artificial as there is no leakage occurring.
